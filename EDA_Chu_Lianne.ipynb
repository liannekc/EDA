```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv("MA_Public_Schools_2017.csv")
df.head()
```

```
# filter out the high schools
df_hs = df[df['12_Enrollment']>0]
df_hs.head()
```

Q1: What factors affect SAT scores?

```
# create 3 subplots
myFig = plt.figure(figsize=(15, 10))
ax1 = myFig.add_subplot(2, 3, 1)
ax2 = myFig.add_subplot(2, 3, 2)
ax3 = myFig.add_subplot(2, 3, 3)
```

```
sns.boxplot(y="Average SAT_Reading", data=df_hs,ax=ax1)
sns.boxplot(y="Average SAT_Math", data=df_hs,ax=ax2)
sns.boxplot(y="Average SAT_Writing", data=df_hs,ax=ax3)
myFig
```

Not much variation between the 3 different scores so will choose one (Average SAT_Reading) to do all future analysis with

```
# Average Class Size vs Avg SAT Reading
df_hs.plot.scatter("Average Class Size","Average SAT_Reading", title = "Average Class Size vs Average SAT_Reading")
```

Slight positive correlation in the 10-20 class size region

```
# English Language Learner vs Avg SAT Reading
df_hs.plot.scatter("English Language Learner","Average SAT_Reading", title = 'English Language Learner vs Avg SAT Reading' )
```

Larger cluster of data in the 0-50 region. Negative trend afterwards, but very few datapoints

```
# First Language Not English vs Avg SAT Reading
df_hs.plot.scatter("First Language Not English","Average SAT_Reading", title = 'First Language Not English vs Avg SAT Reading' )
```

Large cluster of data in 0-100 region

```
# % of High need students vs AVG SAT Reading
df_hs.plot.scatter("% High Needs","Average SAT_Reading")
```

Strong negative correlation

```
# Average Expenditure per pupil vs AVG SAT Reading
df_hs.plot.scatter("Average Expenditures per Pupil","Average SAT_Reading", title = 'Average Expenditure per pupil vs AVG SAT Reading')
```

No clear correlation

```
# Average Teacher Salary vs Avg SAT Reading
df_hs.plot.scatter("Average Salary","Average SAT_Reading", title='Average Teacher Salary vs Avg SAT Reading')
```

Moderate positive correlation

```
# School Accountability Percentile vs Avg SAT Reading
df_hs.plot.scatter("School Accountability Percentile (1-99)","Average SAT_Reading")
```

Strong positive correlation

```
# % of Economically Disadvantged vs Avg SAT Reading
df_hs.plot.scatter("% Economically Disadvantaged","Average SAT_Reading",title='% of Economically Disadvantged vs Avg SAT Reading')
```

Strong negative correlation

The different factors that have a strong correlation to Average SAT_Reading Scores are:
- % High Needs (number of students with high needs in the school)
- Average Salary (average salary of the teacher)
- School Accountability Percentile
- % Economically Disadvatanged

There seems to be a strong negative correlation with % of students that are econmically disadvantaged and Average SAT Reading scores. We will then use % economically disadvantaged to further explore the data

Q2: Does higher SAT score lead to higher college attendance?





```
# % Attending College vs Average SAT_Reading score
df_hs.plot.scatter("Average SAT_Reading","% Attending College",title='% Attending College vs Average SAT_Reading score')
```

There seems to be positive correlation between Average SAT_ Reading scores and college attendance rate.


Q3: How do the different accountability levels of the school affects its students college attendance percentage for different types of colleges?


```
# Box plot of different accountability levels in relation to % of economically disadvantaged
sns.boxplot(x="Accountability and Assistance Level", y="% Economically Disadvantaged", data=df_hs)
```

Level 1 schools have the least economically disadvantaged students and Level 5 has the most. Because the Accountabilioty and Assistance level is categorical data, we will use that for subsequent graphs instead of "% of Economically Disadvantaged"

```
# Average SAT_Reading vs % Private Two-Year
sns.lmplot('Average SAT_Reading', '% Private Two-Year', data=df_hs, hue='Accountability and Assistance Level', fit_reg=False).set(title='Average SAT_Reading vs % Private Two-Year')
```

Many null values (not many students attend private two year colleges, hence will not put weight on outcomes from the graph

```
# Average SAT_Reading vs % MA Community College
sns.lmplot('Average SAT_Reading', '% MA Community College', data=df_hs, hue='Accountability and Assistance Level', fit_reg=False).set(title='Average SAT_Reading vs % MA Community College')
```

Negative correlation

Students with lower SAT scores choose to attend Massachusetts community college. Students with higher SAT scores opt for other college options

```
# Average SAT_Reading vs % Public Two-Year
sns.lmplot('Average SAT_Reading', '% Public Two-Year', data=df_hs, hue='Accountability and Assistance Level',fit_reg=False).set(title='Average SAT_Reading vs % Public Two-Year')
```

Negative correlation

Similar outcome to community college graph

```
# Average SAT_Reading vs % Public Four-Year
sns.lmplot('Average SAT_Reading', '% Public Four-Year', data=df_hs, hue='Accountability and Assistance Level', fit_reg=False).set(title='# Average SAT_Reading vs % Public Four-Year')
```

Positive correlation

As student's SAT scores increase, so does their public 4 year college attendance rate. This positive correlation shows that public 4 years are more competitive than public 2 year colleges

```
# Average SAT_Reading vs % Private Four-Year
sns.lmplot('Average SAT_Reading', '% Private Four-Year', data=df_hs, hue='Accountability and Assistance Level', fit_reg=False).set(title='Average SAT_Reading vs % Private Four-Year')
```

Strong Positive correlation

Similar to above, as student's SAT scores increase, so does their private 4 year college attendance rate.

Level 1 schools have higher attendance percentage, which makes sense because private colleges tend to be more expensive. Families with a higher socioecomic background most likely attend higher performing high schools

```
myFig2 = plt.figure(figsize=(15, 10))
ax1 = myFig2.add_subplot(2, 3, 1)
ax2 = myFig2.add_subplot(2, 3, 2)
ax3 = myFig2.add_subplot(2, 3, 3)
```
